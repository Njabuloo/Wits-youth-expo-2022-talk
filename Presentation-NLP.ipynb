# NLP - Natural Langauge Processing

## What is NLP?
- NLP is a branch of Artficial Intelligence that deals with the ability of computers being able to understand human language as its spoken and written.

## NLP Examples 
- Sentiment Analysis
- Topic Modeling
- Text Generation

### Sentiment Analysis
- An NLP technique used to determine whether data is positive, negative or neutral.
- Example: used in call centers, chat bots, social media to analyze how people feel about a product.

### Topic Modeling
- A NLP technique used to extract words and phrases within a document, and classify them into groupings, depending to their related expressions. 
- Example : Spam dectention in emails.

### Text Generation
- A sub-field of NLP.
 It is used to generate text that appears to be indistinguishable from human written text.
- Example : An AI model that writes essays : https://www.the-good-ai.com/

### 3. Listening
- Our virtual assitant can listen to what we say to it.

```python
# Object that will be able to recognize our speech
listener = sr.Recognizer()

def listen():
    #test if the mictophone works or not
    try :
        with sr.Microphone() as source:
            print("Listening.")
            #save the speech that is being read in
            voice = listener.listen(source)
            #print the text to check if it can listen perfectly file
            text = listener.recognize_google(voice)
            print(text)
    #throws an exception error if the microphone is not working
    except :
        pass
```

```python
listen()
```

### 4. Talk back
- Our virtual assistant can talk back

```python
#initialize the engine to be able to talk back
engine = pyttsx3.init()

#can say any speech
def talk(speech):
    engine.say(speech)
    engine.runAndWait()
```

```python
speech = 'Hey, I am Sally, your virtual assistant'
talk(speech)
```
